#### Imports

```python
import numpy as np
import scipy.stats 
import matplotlib.pyplot as plt
```

#### Plot Likehood

```python
n = 10                                       # set total trials
s = 8                                        # set total successes

theta = np.linspace(start= 0, stop = 1, num = 100) # create theta variable from 0 to 1

like = scipy.stats.binom(n, theta).pmf(k = s)      # create likelihood function

plt.plot(theta, like)
plt.xlabel('theta')
plt.ylabel('Likelohood')
plt.title('Likelohood Curve')
plt.show()
```

#### Calculate Likelihood Ratio

```python
n = 13                                         # set total trials
s = 8                                          # set successes

H0 = 0.5                                       # specify one hypothesis you want to compare with the likelihood ratio
H1 = 1/20                                      # specify second hypothesis you want to compare with the likelihood ratio

like_H0_H1 = scipy.stats.binom(n, H0).pmf(k = s)/scipy.stats.binom(n, H1).pmf(k = s) # likelihood ratio of H0 over H1
like_H1_H0 = 1/like_H0_H1                                                            # likelihood ratio of H1 over H0

theta = np.linspace(start = 0, stop = 1, num = 50)
like = scipy.stats.binom(n, theta).pmf(k = s)      # create likelihood function

plt.figure(figsize = (14,5))
plt.plot(theta, like)                              # plot the likelihood curve
plt.plot([H0],[scipy.stats.binom(n, H0).pmf(k = s)], marker = 'o') # likelihood of H0
plt.plot([H1],[scipy.stats.binom(n, H1).pmf(k = s)], marker = 'o') # likelihood of H1
plt.plot([H0, s/n],[scipy.stats.binom(n, H0).pmf(k = s),scipy.stats.binom(n, H0).pmf(k = s)], ls = '--', color = 'orange')
plt.plot([H1, s/n],[scipy.stats.binom(n, H1).pmf(k = s),scipy.stats.binom(n, H1).pmf(k = s)], ls = '--', color = 'orange')
plt.plot([s/n, s/n],[scipy.stats.binom(n, H0).pmf(k = s),scipy.stats.binom(n, H1).pmf(k = s)], ls = '-', color = 'black')
plt.xlabel('theta')
plt.ylabel('Likelihood')
plt.title('Likelihood ratio H0/H1: {0:2f} \nLikelihood ratio H1/H0: {1:e}'.format(like_H0_H1, like_H1_H0))
plt.show()
```
